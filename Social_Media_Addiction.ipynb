# Academic Impact

```python
df.head(2)
```

```python
sns.barplot(x="Affects_Academic_Performance",y="Addicted_Score",hue="Gender",data=df)
plt.xlabel("Effect on Academic Performance")
plt.title("Relation between Academic Performance and Score of Addiction")
```

There is a clear relationship between social media usage and academic performance. Students who report that social media negatively affects their academic results tend to have higher addiction scores, typically between 7 and 8. In contrast, those whose academic performance remains unaffected generally have lower addiction scores around 4 to 5. This suggests that increased social media dependence may be linked to poorer academic outcomes.

```python
sns.boxplot(data=df, x='Affects_Academic_Performance', y='Avg_Daily_Usage_Hours', palette="Set2")
plt.title('Avg Daily Usage vs Academic Performance Impact')
plt.xlabel('Affects Academic Performance?')
plt.ylabel('Average Daily Usage (Hours)')
plt.tight_layout()
plt.show()
```

for higher average usage hours , academic performance is negatively affected.

```python
platform_impact = df.groupby(['Most_Used_Platform', 'Affects_Academic_Performance']).size().unstack()

platform_impact.plot(kind='bar', stacked=True, figsize=(10, 6), colormap='tab20c')
plt.title('Academic Impact by Most Used Social Media Platform')
plt.xlabel('Platform')
plt.ylabel('Number of Students')
plt.legend(title='Affects Academic Performance')
plt.tight_layout()
plt.show()
```

Instagram, Facebook, and TikTok are the most frequently used apps and have a significant negative impact on students' academic performance. In contrast, apps like Line, VKontakte, and LinkedIn, which are used less often, show minimal effect on academic outcomes. Although Snapchat is less popular among students, its high addiction potential also contributes to a negative impact on their academic performance.

```python
sns.barplot(x="Affects_Academic_Performance",y="Sleep_Hours_Per_Night",hue="Addicted_Score",data=df)
plt.title("Relation between Academic Performance and Sleep Hours")
plt.xlabel("Effect on Academic Performance")
plt.ylabel("Sleep Hours")
```

There is a clear connection between sleep duration and social media addiction. Students who sleep fewer hours tend to have higher addiction scores, indicating greater dependence on social media. Conversely, students who maintain healthy sleep patterns show lower addiction levels, and their academic performance remains unaffected.

```python
plt.figure(figsize=(10, 6))
sns.barplot(x="Addicted_Score",y="Sleep_Hours_Per_Night",hue="Affects_Academic_Performance",data=df)
plt.title("Relation between Addiction and Sleep Hours")
plt.xlabel("Addicted Score")
plt.ylabel("Sleep Hours")
```

```python
sns.barplot(x="Conflicts_Over_Social_Media",y="Mental_Health_Score",hue="Affects_Academic_Performance",data=df)
plt.title("Relation between Conflicts and Mental Health")
plt.xlabel("Conflicts Over Social Media")
plt.ylabel("Mental Health Score")
```

Students who experience more conflicts related to social media tend to have lower mental health scores and significantly poorer academic performance, which is a serious concern. In contrast, those with fewer social media-related conflicts generally maintain normal mental health scores around 8 to 9 and better academic outcomes. This highlights a strong correlation between social media conflicts, mental well-being, and academic performance.

```python
platform_metrics = df.groupby("Most_Used_Platform")[[
    "Addicted_Score",
    "Avg_Daily_Usage_Hours",
    "Mental_Health_Score",
    "Conflicts_Over_Social_Media",
    "Sleep_Hours_Per_Night"
]].mean()

# ---- Heatmap 1: Risky Behaviors (high = bad) ----
risk_metrics = platform_metrics[[
    "Addicted_Score",
    "Avg_Daily_Usage_Hours",
    "Conflicts_Over_Social_Media"
]]
risk_normalized = (risk_metrics - risk_metrics.min()) / (risk_metrics.max() - risk_metrics.min())
risk_sorted = risk_normalized.sort_values("Avg_Daily_Usage_Hours", ascending=False)

plt.figure(figsize=(10, 5))
sns.heatmap(risk_sorted, annot=True, cmap="YlOrRd", linewidths=0.5)
plt.title("Risky Platform Behaviors (Higher is Worse)")
plt.xlabel("Behavioral Metric")
plt.ylabel("Most Used Platform")
plt.tight_layout()
plt.show()

# ---- Heatmap 2: Wellbeing Behaviors (high = good) ----
wellbeing_metrics = platform_metrics[[
    "Mental_Health_Score",
    "Sleep_Hours_Per_Night"
]]
wellbeing_normalized = (wellbeing_metrics - wellbeing_metrics.min()) / (wellbeing_metrics.max() - wellbeing_metrics.min())
wellbeing_sorted = wellbeing_normalized.sort_values("Mental_Health_Score", ascending=True)

plt.figure(figsize=(8, 5))
sns.heatmap(wellbeing_sorted, annot=True, cmap="YlGnBu", linewidths=0.5)
plt.title("Positive Platform Outcomes (Higher is Better)")
plt.xlabel("Wellbeing Metric")
plt.ylabel("Most Used Platform")
plt.tight_layout()
plt.show()
```
